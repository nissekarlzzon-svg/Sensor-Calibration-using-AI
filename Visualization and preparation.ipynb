```python
def pm10_by_station(station_code, date_range=None):
    pm10 = Measurement_info[
    (Measurement_info['Item code'] == 8) &   # PM10
    (Measurement_info['Instrument status'] == 0)
    ].copy()

    station = pm10[pm10['Station code'] == station_code].copy()

    station['Measurement date'] = pd.to_datetime(station['Measurement date'])

    station = station.sort_values('Measurement date')

    if date_range:
        station = station[
            (station['Measurement date'] >= date_range[0]) &
            (station['Measurement date'] <= date_range[1])
        ]

    plt.figure(figsize=(15,5))
    plt.plot(
    station['Measurement date'],
    station['Average value'],
    linewidth=0.7
    )

    plt.title(f"PM10 over time – Station {station_code}")
    plt.xlabel("Date")
    plt.ylabel("PM10 (µg/m³)")
    plt.axhline(30, color='blue', linestyle='--', alpha=0.6, label='Good (30)')
    plt.axhline(80, color='green', linestyle='--', alpha=0.6, label='Normal (80)')
    plt.axhline(150, color='orange', linestyle='--', alpha=0.6, label='Bad (150)')
    #plt.axhline(600, color='red', linestyle='--', alpha=0.6, label='Very bad (600)')
    plt.axhspan(0, 30, color='blue', alpha=0.2)
    plt.axhspan(30, 80, color='green', alpha=0.2)
    plt.axhspan(80, 150, color='orange', alpha=0.2)
    plt.axhspan(150, 400, color='red', alpha=0.2)
    plt.grid(True)
    plt.show()
```

Station 119 is the most polluted station from pm10, thus making it the most intresting station. For that reason it is the one we will be focusing on fromm now on. Station 119 has the latitude 37.525007 and the longtitude 126.897370. From these values we can extract the moisture and temperatures for the relevant time-period from https://open-meteo.com/en/docs/historical-weather-api. These values can then be used to simulate a sensor that is very sensitive to these factors to create a pm10_raw. The initial data will be used as true values simulating a reference-grade sensor.

```python
# this code cell is from the Open-Meteo API tutorial: https://open-meteo.com/en/docs#api-formats
#The relevant dates and the latitude and longtitude for station 119 are inserted

import openmeteo_requests

import pandas as pd
import requests_cache
from retry_requests import retry

# Setup the Open-Meteo API client with cache and retry on error
cache_session = requests_cache.CachedSession('.cache', expire_after = -1)
retry_session = retry(cache_session, retries = 5, backoff_factor = 0.2)
openmeteo = openmeteo_requests.Client(session = retry_session)

# Make sure all required weather variables are listed here
# The order of variables in hourly or daily is important to assign them correctly below
url = "https://archive-api.open-meteo.com/v1/archive"
params = {
	"latitude": 37.525007,
	"longitude": 126.89737,
	"start_date": "2017-01-01",
	"end_date": "2019-12-31",
	"hourly": ["temperature_2m", "relative_humidity_2m"],
}
responses = openmeteo.weather_api(url, params=params)

# Process first location. Add a for-loop for multiple locations or weather models
response = responses[0]
print(f"Coordinates: {response.Latitude()}°N {response.Longitude()}°E")
print(f"Elevation: {response.Elevation()} m asl")
print(f"Timezone difference to GMT+0: {response.UtcOffsetSeconds()}s")

# Process hourly data. The order of variables needs to be the same as requested.
hourly = response.Hourly()
hourly_temperature_2m = hourly.Variables(0).ValuesAsNumpy()
hourly_relative_humidity_2m = hourly.Variables(1).ValuesAsNumpy()

hourly_data = {"date": pd.date_range(
	start = pd.to_datetime(hourly.Time(), unit = "s", utc = True),
	end =  pd.to_datetime(hourly.TimeEnd(), unit = "s", utc = True),
	freq = pd.Timedelta(seconds = hourly.Interval()),
	inclusive = "left"
)}

hourly_data["temperature_2m"] = hourly_temperature_2m
hourly_data["relative_humidity_2m"] = hourly_relative_humidity_2m

hourly_dataframe = pd.DataFrame(data = hourly_data)
print("\nHourly data\n", hourly_dataframe)
hourly_dataframe.head()
```

For each PM10 we now want a factor to multiply with to obtain a value for a simulated, low-grade pm10-sensor. The following formula builds on the assumption that the sensor works as best at 50% humidity and 10 degrees. a and b are sensitivity-factors.

PM10_raw ​= PM10true​*(1+a(H−50)+b(T−10))

Humidity and temperature are not parts of the original dataset and will not be part of the train set. But there may be corrolations in the other factors which may lead to some kind of relationship. That would be too easy so a random, normally distrubuted value is added to the equation as noise.

PM10_raw ​= PM10true​*normRandomValue*(1+a(H−50)+b(T−10))

```python
summary_119['Measurement date'] = pd.to_datetime(summary_119['Measurement date'])
hourly_dataframe['date'] = pd.to_datetime(hourly_dataframe['date']).dt.tz_localize(None)

summary_119 = summary_119.merge(
    hourly_dataframe[['date', 'temperature_2m', 'relative_humidity_2m']],
    left_on='Measurement date',
    right_on='date',
    how='left'
)

def add_PM10_raw(PM10, temp, humidity, a=0.05,b=0.02):
    randomnoise = np.clip(np.random.normal(1, 0.06, len(PM10)), 0.8, 1.2)
    PM10_raw = PM10 * randomnoise * (1 + a*(humidity - 50) + b*(temp - 10))
    return PM10_raw

summary_119['PM10_raw'] = add_PM10_raw(
    PM10=summary_119['PM10'],
    temp=summary_119['temperature_2m'],
    humidity=summary_119['relative_humidity_2m']
)

summary_119.head()
```
